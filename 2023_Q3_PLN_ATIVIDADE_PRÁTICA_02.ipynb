Aqui estão alguns exemplos de erros comuns que podem ser identificados usando expressões regulares:

* Erros de gramática e ortografia: erros de digitação, concordância verbal e nominal, uso incorreto de pontuação e outros erros gramaticais.

### Função para acionamento das outras

```
from bs4 import BeautifulSoup
import requests
import re
import json
```

```
def Get(item):
  print(f"Iniciando o processamento do capitulo:{item[0]}")
  response = requests.get(item[1])
  soup = BeautifulSoup(response.content, 'html.parser')

  main_element = soup.find('main', {'class': 'content', 'id': 'quarto-document-content'})

  if main_element:
      texto = main_element.get_text().lower()
      textoOffNormalize = main_element.get_text()

  print("Conexão e coleta do texto com sucesso")

  print("\n\nInicio do processamento com Spacy modelo pt_core_news_sm\n\n")
  errosSpacy=SpacyPtCore(texto)
  print("\n\nTermino do processamento com Spacy modelo pt_core_news_sm\n\n")
  print("="*20)
  print("\n\nInicio do processamento com LanguageTooL\n\n")
  errosLanguageTooL = LanguageTooL(textoOffNormalize)
  print("\n\nTermino do processamento com LanguageTooL\n\n")
  print("="*20)
  print("\n\nProcessando erros com Regex\n\n")
  acronimos, abreviacoes, concatenateNumber, palavras_com_espacos_extras, caracteresEspecial, pontoFinal= ErrosWithRegex(texto,textoOffNormalize)
  print(f"Finalizando processamento do capitulo {item[0]} ...")
  print("="*20+"\n\n\n")


  json_data = {
    "Capitulo": item[0],
    "errosSpacy": errosSpacy,
    "errosLanguageTooL": errosLanguageTooL,
    "acronimos": acronimos,
    "abreviacoes": abreviacoes,
    "concatenateNumber": concatenateNumber,
    "palavras_com_espacos_extras": palavras_com_espacos_extras,
    "caracteresEspecial": caracteresEspecial,
    "texto": texto,
    "textoOffNormalize": textoOffNormalize,
    "pontoFinal": pontoFinal
  }

  # Libera a memória
  errosSpacyPtCore = None
  errosSpacy = None
  errosLanguageTooL = None
  acronimos = None
  abreviacoes = None
  concatenateNumber = None
  palavras_com_espacos_extras = None
  caracteresEspecial = None
  texto = None
  textoOffNormalize = None

  return json_data
```

### Modelos para encontrar erros gramáticas

Lembrando que o Spacy não é um modelo voltado para esse tipo de tarefa, portanto, ele pode ter  incongruências indicando falsos positivos,por exemplo, nomes de pessoas, palavras em inglês.

```
 def SpacyPtCore(texto):
  from tqdm import tqdm
  from thinc.layers.concatenate import concatenate
  import spacy
  import re
  # Carregue o modelo de linguagem em português
  nlp = spacy.load("pt_core_news_sm")

  # Vetor para guardar palavras encontradas com erros
  errosSpacy = []

  # Analise o texto
  doc = nlp(texto)

  for token in tqdm(doc, desc="Primeiro processamento de palavras com pt_core_news_sm"):
      # Verifique se a palavra não é uma palavra de parada e não contém números ou caracteres especiais
      if not token.is_stop and not token.is_alpha and not token.is_punct:
          errosSpacy.append(token.text)

  for token in tqdm(doc, desc="Segundo processamento de palavras com pt_core_news_sm"):
      if token.is_alpha and not token.is_oov:
          if not token._.grammar_correct:
              errosSpacy.append(token.text)
  return errosSpacy
```

Utilização do LanguageTool

```
def LanguageTooL(textoOffNormalize):
  import language_tool_python

  tool = language_tool_python.LanguageTool('pt-BR')  # Certifique-se de definir o idioma desejado

  if(len(textoOffNormalize)>5000):
    partes_do_texto = textoOffNormalize.split('\n\n')  # Suponha que os parágrafos sejam separados por duas quebras de linha

    # Lista para armazenar resultados
    erros = []

    # Verifica cada parte do texto
    for parte in partes_do_texto:
        correcoes = tool.check(parte)
        erros.extend(correcoes)
  else:
    erros = tool.check(textoOffNormalize)

  return erros
```

### Encontrando erros com com Regex

```
def ErrosWithRegex(texto,textoOffNormalize):
  # Expressões regulares para encontrar acronimos
  acronimos = re.findall(r'\b[A-Z]+[A-Z]+\b', textoOffNormalize)

  # Expressões regulares para encontrar abreviações
  abreviacoes = re.findall(r'\b[A-Z][a-z]*\.\s*', textoOffNormalize)

  #Expressão regular para encontrar palavras com numeros
  concatenateNumber = re.findall(r'\b\S*\w+\d+\S*\b', texto)
  for token in concatenateNumber:
      try:
          if type(float(token)) == float or (type(int(token)) == int):
              concatenateNumber.remove(token)
      except:
          pass

  # Encontre todas as palavras com espaços extras no texto
  palavras_com_espacos_extras = re.findall(r'\b\w+\s{2,}\w+\b', textoOffNormalize)

  #Expressão regular para encontrar caracteres especiais isolados
  caracteresEspecial= re.findall(r'\b\w+\d+([\W]+)\b', texto)


  #Ponto final
  pontoFinal= re.findall(r'\.$', texto)

  return acronimos, abreviacoes, concatenateNumber, palavras_com_espacos_extras, caracteresEspecial, pontoFinal
```

```
def metricasCaps(listUrl):
    for i in listUrl:
      data = Get(i)

      cap = data["Capitulo"]
      errosSpacy = data["errosSpacy"]
      errosSpacyUnicos = list(set(errosSpacy))

      errosLanguageTooL = data["errosLanguageTooL"]

      acronimos = data["acronimos"]
      acronimosUnicos = list(set(acronimos))

      abreviacoes = data["abreviacoes"]
      abreviacoesUnicos = list(set(abreviacoes))

      concatenateNumber = data["concatenateNumber"]
      concatenateNumberUnicos = list(set(concatenateNumber))

      palavras_com_espacos_extras = data["palavras_com_espacos_extras"]
      palavras_com_espacos_extrasUnicos = list(set(palavras_com_espacos_extras))


      caracteresEspecial = data["caracteresEspecial"]
      caracteresEspecialUnicos = list(set(caracteresEspecial))

      pontoFinal= data["pontoFinal"]



      print(f"\n\nNo capitulo {cap} tem as seguintes métricas:\n\n")
      print(f"O modelo do Spacy encontrou {len(errosSpacy)} erros, sendo {len(errosSpacyUnicos)} únicos. Segue os erros:\n {errosSpacy} \n\n")

      print(f"O modelo do LanguageTool encontrou {len(errosLanguageTooL)} erros.")
      if errosLanguageTooL:
        print("Erros gramaticais e ortográficos encontrados:")
        for erro in errosLanguageTooL:
            print(f"Erro: {erro}")


      print("\n\nUtilizando Regex foi encontrado:\n\n")
      print(f"Foi encontrado {len(acronimos)} acronimos, sendo {len(abreviacoesUnicos)} acronimos únicos. Segue os acronimos:\n {acronimos} \n\n")
      print(f"Foi encontrado {len(abreviacoes)} abreviações, sendo {len(abreviacoesUnicos)} abreviações únicas. Segue as abreviações:\n {abreviacoes} \n\n")
      print(f"Foi encontrado {len(concatenateNumber)} números concatenados com palavras ou caracteres especiais, sendo {len(concatenateNumberUnicos)} ocorrências únicas. Segue todas as ocorrências :\n {concatenateNumber} \n\n")
      print(f"Foi encontrado {len(palavras_com_espacos_extras)} espaços extras, sendo {len(palavras_com_espacos_extrasUnicos)} ocorrências únicas.Segue todas as ocorrências :\n {palavras_com_espacos_extras} \n\n")
      print(f"Foi encontrado {len(caracteresEspecial)} formatação/caracteres especiais, sendo {len(caracteresEspecialUnicos)} ocorrências únicas.Segue todas as ocorrências :\n {caracteresEspecial} \n\n")
      print(f"Foi encontrado {len(pontoFinal)} pontos finais.Segue todas as ocorrências :\n {pontoFinal} \n\n")


```
